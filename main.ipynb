```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error,root_mean_squared_error,adjusted_rand_score
from sklearn.linear_model import Ridge,Lasso
from sklearn.datasets import load_diabetes
```

```python
diabetis = load_diabetes()
```

```python
diabetis
```

```python
df  = pd.DataFrame(diabetis.data, columns=diabetis.feature_names)
```

```python
df['target'] = diabetis.target
```

```python
df
```

```python
x = df.drop(columns=['target'])
```

```python
x
```

```python
y = df['target']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
scalar = StandardScaler()
```

```python
x_train_scaled = scalar.fit_transform(x_train)
```

```python
x_test_scaled = scalar.transform(x_test)
```

```python
Ridge_model = Ridge()
Lasso_model  = Lasso()
```

```python
Ridge_model.fit(x_train_scaled, y_train)
```

```python
Lasso_model.fit(x_train_scaled, y_train)
```

```python
Ridge_model.predict(x_test_scaled)
```

```python
Lasso_model.predict(x_test_scaled)
```

```python
mean_squared_error(y_test, Ridge_model.predict(x_test_scaled))
```

```python
mean_squared_error(y_test, Lasso_model.predict(x_test_scaled))
```

```python
mean_absolute_error(y_test, Ridge_model.predict(x_test_scaled))
```

```python
mean_absolute_error(y_test, Lasso_model.predict(x_test_scaled))
```

```python
r2_score(y_test, Ridge_model.predict(x_test_scaled))
```

```python
r2_score(y_test, Lasso_model.predict(x_test_scaled))
```

```python
root_mean_squared_error(y_test, Ridge_model.predict(x_test_scaled))
```

```python
root_mean_squared_error(y_test, Lasso_model.predict(x_test_scaled))
```
